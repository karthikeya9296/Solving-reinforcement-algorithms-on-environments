```python
import pickle
import matplotlib.pyplot as plt

def plot_rewards(modified_env, vanilla_env):
  plt.plot(modified_env, label='Noise Environment Reward')
  plt.plot(vanilla_env, label='Vanilla Environment Reward')
  plt.xlabel('Episode')
  plt.ylabel('Reward Value')
  plt.title('A3C on Cart Pole Environment')
  plt.grid()
  plt.legend()
  plt.show()

def read_rewards(filename):
  with open(filename, 'rb') as file:
    return pickle.load(file)

def plot_rewards_algo(dqn, ddqn, a2c, a3c):
  plt.plot(dqn, label='DQN')
  plt.plot(ddqn, label='DDQN')
  plt.plot(a2c, label='A2C')
  plt.plot(a3c, label='A3C')
  plt.xlabel('Episode')
  plt.ylabel('Reward Value')
  plt.title('Cart Pole Noise Environment')
  plt.grid()
  plt.legend()
  plt.show()
```

#DQN Cart Pole Vanilla vs Noise Environment

```python
# Train Rewards

vanilla_rewards = read_rewards('dqn_vanilla_cartpole_train.pkl')
modified_rewards = read_rewards('dqn_noise_cartpole_train.pkl')

plot_rewards(modified_rewards, vanilla_rewards)
```

```python
# Test Rewards

vanilla_rewards = read_rewards('dqn_vanilla_cartpole_test.pkl')
modified_rewards = read_rewards('dqn_noise_cartpole_test.pkl')

plot_rewards(modified_rewards, vanilla_rewards)
```
